### Transcribing DNA into RNA
Problem
An RNA string is a string formed from the alphabet containing 'A', 'C', 'G', and 'U'.

Given a DNA string t
 corresponding to a coding strand, its transcribed RNA string u
 is formed by replacing all occurrences of 'T' in t
 with 'U' in u
.

Given: A DNA string t
 having length at most 1000 nt.

Return: The transcribed RNA string of t
.

image.png


```python
def transcribed_rna(original_rna):
  rna = original_rna.lower()
  result = ''
  for i in rna:
    if i == 't':
      result += 'u'
    else:
      result += i
  return result.upper()
q2_1 = 'GATGGAACTTGACTACGTAAATT'
test_q2_1 = transcribed_rna(q2_1)
print(test_q2_1)
```

### Complementing a Strand of DNA

```python
#q3
def reverse_compliment(dna):
  dna = dna.lower()
  result = ''
  reverse_dna = dna[::-1]
  for i in reverse_dna:
    if i == 'a':
      result += 'T'
    elif i == 't':
      result += 'A'
    elif i == 'c':
      result += 'G'
    else:
      result += 'C'
  return result


q3_1 = 'AAAACCCGGT'
test_q3_1 = reverse_compliment(q3_1)
print(test_q3_1)
```

DNA strings must be labeled when they are consolidated into a database. A commonly used method of string labeling is called FASTA format. ### Hamming distance
Problem

Figure 2. The Hamming distance between these two strings is 7. Mismatched symbols are colored red.
Given two strings s
 and t
 of equal length, the Hamming distance between s
 and t
, denoted dH(s,t)
, is the number of corresponding symbols that differ in s
 and t
. See Figure 2.

Given: Two DNA strings s
 and t
 of equal length (not exceeding 1 kbp).

Return: The Hamming distance dH(s,t)
.

```python
def hamming_distance(s,t):
  count = 0
  s = s.lower()
  t = t.lower()
  for i in range(len(s)):
    if s[i] != t[i]:
      count += 1
  return count
s1 = 'GAGCCTACTAACGGGAT'
t1 = 'CATCGTAATGACGGCCT'
print(hamming_distance(s1,t1))
```

### Mendel's first law

```python
#P(Aa) = P(A|a) + P(a|A)
#P(A|a)
def mendel(k,m,n):
  """k: homozygous dominant, m: hetero, n: homoz recessive"""
  #P(A|a)
  #When we choose a => the other must be A
  #P(a)
  total = k + m + n
  tot = total - 1#total minus 1
  #pAa = 1/2*m/total*(k/total_minus_1 + (m-1)/total_minus_1*1/2) + n/total*(1/2*m/total_minus_1 + k/total_minus_1)
  #paA = k/total*(n/total_minus_1 + 1/2*m/total_minus_1) + 1/2*m/total*((m-1)*1/2/total_minus_1 + n/total_minus_1)
  pAa = 1/2*m/total*((m-1)/2*(1/tot) + k/tot) + n/total*(m/2*(1/tot) + k/tot)
  paA = m/2*(1/total)*((m-1)/2*(1/tot) + n/tot) + k/total*(m/2*(1/tot)+n/tot)
  pAA = k/total*((m/tot)*1/2 + (k-1)/tot) + 1/2*m/total*(k/tot + (m-1)/tot*1/2)
  return pAa + paA + pAA

print(mendel(2,2,2))
```

```python
#Fibonaci using dynamic programming
def Fibo(n):
  seq = np.zeros(n)
  seq[0] = 1
  seq[1] = 1
  for i in range(2,n):
    seq[i] = seq[i - 1] + seq[i - 2]
  return seq[n-1]
print(Fibo(11))
```

```python
#Immortal rabbit
def Rabit_immortal(n,m):
  seq = np.zeros(n)
  seq[0] = 1
  seq[1] = 1
  if n <= m:
    for i in range(2,n):
      seq[i] = seq[i - 1] + seq[i - 2]
    return seq[n-1]
  else:
    for i in range(2,n):
      if i > m:
        seq[i] = seq[i - 1] + seq[i - 2]
      else:
        seq[i] = seq[i - 1] + seq[i - 2] - seq[i - (m + 1)]
    return seq[n-1]
print(Rabit_immortal(6,3))
```

```python
def Rabbit_immortal(n, m):
    # Initialize a list to track the number of rabbits at each age
    rabbits = [1, 1]

    # Initialize a variable to keep track of the number of months
    current_month = 2

    # Continue calculating rabbit population until reaching age 'n'
    while current_month < n:
        if current_month < m:
            # If rabbits are younger than 'm', they reproduce normally
            new_rabbits = rabbits[-2] + rabbits[-1]
        elif current_month == m or current_month == m + 1:
            # If rabbits are 'm' or 'm + 1' months old, account for mortality
            new_rabbits = rabbits[-2] + rabbits[-1] - 1
        else:
            # For all other months, consider mortality and reproduction
            new_rabbits = rabbits[-2] + rabbits[-1] - rabbits[-(m + 1)]

        # Add the newly calculated rabbit population to the list
        rabbits.append(new_rabbits)

        # Increment the number of months
        current_month += 1

    # Return the number of rabbits at age 'n'
    return rabbits[-1]
print(Rabbit_immortal(6,3))
```

### Independent Alleles

```python
def factorial(n):
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def combinations(n, k):
    if k < 0 or k > n:
        return 0
    return factorial(n) // (factorial(k) * factorial(n - k))

def calculate_probability(k, N):
    total_generations = 0
    current_generation = [1.0]  # Initial generation with one organism
    probability = 1.0
    N_count = 0

    while total_generations < k:
        next_generation = []
        for prob in current_generation:
            next_generation.extend([prob * 1/4, prob * 1/4])  # Two offspring per organism
        probability = 1/4  # Probability of AaBb genotype in offspring
        total_generations += 1
        current_generation = next_generation

    at_least_N_probability = 0

    for k_organisms in range(N, len(current_generation) + 1):
        probability_k = combinations(len(current_generation), k_organisms) * (probability ** k_organisms) * ((1 - probability) ** (len(current_generation) - k_organisms))
        at_least_N_probability += probability_k

    return at_least_N_probability

# Example usage:
k = 5
N = 9
result = calculate_probability(k, N)
print(round(result,3))
```

```python
rosalind_revp = '/content/drive/MyDrive/Colab Notebooks/Rosalind/rosalind_revp.txt'
with open(rosalind_revp, 'r') as file:
    lines = file.readlines()
def get_sequence(sep):
  sep = "".join(sep)#because the sep is a list
  return sep
# Define a regular expression pattern to match "Rosalind_xxxx" where 'xxxx' is a number between 0000 and 9999
pattern = r'>Rosalind_(\d{1,4}|[0-9]{1,4}|[0-9]{1,3}[0-9]{1,3}|[0-9]{1,2}[0-9]{1,2}[0-9]{1,2}|[0-9]{1,1}[0-9]{1,1}[0-9]{1,1}[0-9]{1,1})\n'
all_labels = []
for line in lines:
    if re.match(pattern, line):
        all_labels.append(line)

#seperate
lines_1 = lines.copy()
dna_string = []

for j in range(len(all_labels)):

    sep = []
    for i in range(1,len(lines_1)):
      if lines_1[i] not in all_labels:
        sep.append(lines_1[i].strip())
      else:
        end = i
        del lines_1[0:i]
        break
    dna_string.append([all_labels[j][1:].rstrip(),get_sequence(sep)])#get the label with its corresponding sequence
dna_string

#idea create a tuple (pos, Xi, length, reverse_complement) => compare Xi and reverse
dna = dna_string[0][1]
the_dna_tuple = []
all_kmers = []
k = 4
while k < 13:
  for i in range(len(dna)-k+1):
    temp = dna[i:i+k]
    length = len(temp)
    temp_complement = ''
    for j in temp:
      if j == 'A':
        temp_complement += 'T'
      elif j == "T":
        temp_complement += 'A'
      elif j == "G":
        temp_complement += 'C'
      else:
        temp_complement += 'G'
    reverse_complement = temp_complement[::-1]
    if temp == reverse_complement:
      the_dna_tuple.append([i+1,temp, length, temp_complement, reverse_complement])
      all_kmers.append([i+1, length])
  k += 1
  #We will use bubble sort here (the position)
def swap(i,j):
  flag = False#if swap happen => True
  if i > j:
    temp = i
    i = j
    j = temp
    flag = True
  return i,j,flag
def sorting(dna_strings):
  k = len(dna_strings)-1
  while k > 0:
    for i in range(k):
      #dna_strings[i][0],dna_strings[i+1][0] = swap(dna_strings[i][0],dna_strings[i+1][0])
      flag = False
      _,__, flag = swap(dna_strings[i][0],dna_strings[i+1][0])
      if flag:
        dna_strings[i],dna_strings[i+1] = dna_strings[i+1],dna_strings[i]
      #print(dna_strings)
    k -= 1

  return dna_strings
result = sorting(all_kmers)
for i in result:
  for j in i:
    print(j,"",end='')
  print("")
```
